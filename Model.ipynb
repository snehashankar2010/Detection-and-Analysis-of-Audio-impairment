# Model Choosing and Development

```python
#importing/loading libraries
import pandas as pd
import numpy as np 

df1 = pd.read_csv("train.csv",index_col=False)
df2 = pd.read_csv("test.csv",index_col=False)
df1.drop('Unnamed: 0',axis=1, inplace=True)
df2.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
#splitting into x_test,x_train,y_test,y_train
array1 = df1.values
x_train = array1[:,:-2]
y_train = array1[:,-1]

array2 = df2.values
x_test = array2[:,:-2]
y_test = array2[:,-1]
```

```python
#importing models:
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import CategoricalNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
classifiers = []

model1 = SVC()
classifiers.append(model1)

model2 = MLPClassifier(max_iter = 700, activation = 'relu')
classifiers.append(model2)

model3 = KNeighborsClassifier()
classifiers.append(model3)

model4 = LogisticRegression(solver='lbfgs', max_iter=700)
classifiers.append(model4)

model5 = DecisionTreeClassifier(criterion='entropy')
classifiers.append(model5)
```

```python
for clf in classifiers:
    clf.fit(x_train, y_train)
    y_pred= clf.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy of %s is %s"%(clf, acc))
    cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
    print("Confusion Matrix of %s is %s"%(clf, cm))
    print("\n")
```

```python
#finalising the model as logistic regression
model = MLPClassifier(max_iter = 700, activation = 'relu')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
```
